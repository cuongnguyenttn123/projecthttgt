```python
import urllib.request
from bs4 import BeautifulSoup
import csv
```

```python
# specify the url
urlpage =  'https://cacnuoc.vn/tin/top-100-nguoi-giau-nhat-the-gioi/'
# query the website and return the html to the variable 'page'
page = urllib.request.urlopen(urlpage)
# parse the html using beautiful soup and store in variable 'soup'
soup = BeautifulSoup(page, 'html.parser')
print (soup)
```

```python
# find results within table
tbody = soup.find('tbody', attrs={'id': 'list-table-body'})
print(tbody)
```

```python
results = tbody.find_all('tr')
print('Number of results', len(results))
```

```python
# create and write headers to a list 
rows = []
rows.append(['rank', 'name', 'totalassets', 'age', 'company', 'national'])
print(rows)
```

```python
count = 0
for result in results:
    # find all columns per result
    data = result.find_all('td')
    # check that columns have data 
    if len(data) == 1: 
        continue
    # write columns to variables
    rank = data[1].getText()
    ten = data[2].getText()
    tongtaisan = data[3].getText()
    tongtaisan = tongtaisan.split(',')
    if(len(tongtaisan)==2):
        tongtaisan = tongtaisan[0]+'.'+tongtaisan[1]
        tongtaisan = float(tongtaisan)
    else:
        tongtaisan = tongtaisan[0]
    tuoi = data[4].getText()
    st = '-'
    if(tuoi ==st):
        print(tuoi)
    nguon = data[5].getText()
    quoctich = data[6].getText()
    rows.append([rank, ten, tongtaisan, tuoi, nguon, quoctich])
# Create csv and write rows to output file
with open('top100people12345.csv','w', encoding='utf-8', newline='') as f_output:
    csv_output = csv.writer(f_output)
    csv_output.writerows(rows)
```
